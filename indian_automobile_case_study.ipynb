```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
```

```python
dataset = pd.read_csv("Indian automoble buying behavour study 1.0.csv")
dataset
```

```python
dataset.isna().sum()
```

```python
dataset.dtypes
```

```python
dataset['Age'] = dataset['Age']
dataset['No of Dependents'] = dataset['No of Dependents']
dataset['Salary'] = dataset['Salary']
dataset['Wife Salary'] = dataset['Wife Salary']
dataset['Total Salary'] = dataset['Total Salary']
dataset['Price'] = dataset['Price']

dataset['Profession'] = dataset['Profession'].map({'Salaried': 1, 'Business': 0})
dataset['Marrital Status'] = dataset['Marrital Status'].map({'Single': 1, 'Married': 0})
dataset['Education'] = dataset['Education'].map({'Post Graduate': 1, 'Graduate': 0})
dataset['Personal loan'] = dataset['Personal loan'].map({'Yes': 1, 'No': 0})
dataset['House Loan'] = dataset['House Loan'].map({'Yes': 1, 'No': 0})
dataset['Wife Working'] = dataset['Wife Working'].map({'Yes': 1, 'No': 0})

make_mapping = {'i20': 1, 'Ciaz': 2, 'Duster': 3, 'City': 4, 'SUV': 5, 'Creata': 6}
dataset['Make'] = dataset['Make'].map(make_mapping)

dataset = dataset.apply(pd.to_numeric, errors='coerce').astype('float')
dataset.dropna(inplace=True)
dataset.dtypes
```

```python
X = dataset.drop(columns="Price")
X
```

```python
y = dataset["Price"]
y
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace=True)
dataset.isna().sum()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
predictions = model.predict(X_test)
r2_score = r2_score(y_test, predictions)
r2_score
```
